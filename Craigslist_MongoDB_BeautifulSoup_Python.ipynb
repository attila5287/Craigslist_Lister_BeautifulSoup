```python
def Craigslist_MongoDB_BeautifulSoup_Python(item,count):
    pass
    url = 'https://denver.craigslist.org/search/sss?sort=rel&query=' + item
    response = requests.get(url)
    adindex = 0
    soup = BeautifulSoup(response.text, 'lxml')
    results = soup.find_all('li', class_='result-row')
    
    
    # Loop through returned results
    for result in results[:count]:
        # Error handling
        try:
            # Identify and return title of listing
            title = result.find('a', class_='result-title').text
            # Identify and return price of listing
            price = result.a.span.text
            # Identify and return link to listing
            link = result.a['href']
            
            # Run only if title, price, and link are available
            if (title and price and link):

                # Print results
                print(adindex+1,'- - - - - - - - - - - - - - - - - - - - - - ')
                print(title)
                print(price)
                print(link)
                adindex+=1
            # - - - 
                
            
            # Dictionary to be inserted as a MongoDB document
                post = {
                    'adindex':adindex,
                    'title': title,
                    'price': price,
                    'url': link
                }

                collection.insert_one(post)
    
        except Exception as e:
            print(e)
        
    # MongoDB objects reference variable
    listings = db.items.find()
    # - - displayed
    for listing in listings[:count]:
        print(listing)
    
    # save an output file of html            
    with open(results_soup_output,'w') as txt_only:
        for i in range(1,len(results)):
            txt_only.writelines(str(results[i]))
# - - - - - - - 
# - - - - - - - 
# TEST_FUNCTION
Craigslist_MongoDB_BeautifulSoup_Python('guitar',20)
```
